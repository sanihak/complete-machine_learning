```python
#determine the distribution of two classes
credit_dataset['Class'].value_counts()
```

```python
#this is an imbalnced dataset
#0 = represents the transations are legal
#1 =represents fraudulent transactions
```

```python
#seprating the legit and fraudulant transactions

legit =credit_dataset[credit_dataset.Class == 0]
fraud = credit_dataset[credit_dataset.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#undersampling = which is a method to handle imbalance data
#building a sample dataset containing similiar distribution of legit and fraudulent transactions
```

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

```python
#concatinate the two dataframes
```

```python
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```
